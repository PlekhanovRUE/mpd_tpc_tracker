```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot(infile, param_name):
    try:
        with open(infile, encoding='utf-8') as data:
            param_info = json.load(data)
    except Exception as err:
        print('Cannot read json-file')
        return
    df_param = pd.DataFrame(columns=['duration(min)', 'memory', 'eff_sel', 'fake_sel', param_name])
    for _, v in param_info.items():
        dur = v['duration(min)']
        mem = v['memory']
        eff_sel = v['eff_sel']
        fake_sel = v['fake_sel']
        params = v['params']
        param = params[param_name]

        df_param.loc[len(df_param.index)] = dur, mem, eff_sel, fake_sel, param

    df_param.plot(x=param_name, y='eff_sel', kind='scatter')
    plt.show()

    df_param.plot(x=param_name, y='fake_sel', kind='scatter')
    plt.show()

    df_param.plot(x=param_name, y='duration(min)', kind='scatter')
    plt.show()

    df_param.plot(x=param_name, y='memory', kind='scatter')
    plt.show()
```

```python
plot('../../../mpd_tpc_tracker/log_params/maxptscattering.json', "'MaxPtScattering'")
```
